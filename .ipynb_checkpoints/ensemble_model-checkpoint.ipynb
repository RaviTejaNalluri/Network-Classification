```python
minimum_count = 5000
data = data.groupby('Label').apply(lambda x: x.sample(minimum_count) if len(x) >= minimum_count else x).reset_index(drop=True)
```

```python
label_counts = data['Label'].value_counts()
print(label_counts)
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'], unit='s')
data['Start Timestamp'] = pd.to_datetime(data['Start Timestamp'], unit='s')
data['Time Difference'] = data['Timestamp'] - data['Start Timestamp']
```

```python
data = data[['Time Difference', 'TLS Handshake', 'Source IP', 'Destination IP', 'Port', 'Retrain', 'Label']]
```

```python
data['Time Difference'] = data['Time Difference'].dt.total_seconds()
```

```python
data['TLS Handshake'] = data['TLS Handshake'].map({'No': 0, 'Yes': 1})
```

```python
# Encode categorical and textual features
le = LabelEncoder()
data['Source IP'] = le.fit_transform(data['Source IP'])
data['Destination IP'] = le.fit_transform(data['Destination IP'])
data['Label'] = le.fit_transform(data['Label'])
```

```python
X = data.drop(['Label'], axis=1)
y = data['Label']
```

```python
# Split the dataset into training+validation set and test set
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.995, random_state=42)

# Feature scaling and imputation
scaler = StandardScaler()
imputer = SimpleImputer(strategy='mean')

X_train_val_scaled = scaler.fit_transform(X_train_val)
X_train_val_scaled = imputer.fit_transform(X_train_val_scaled)
X_test_scaled = scaler.transform(X_test)
X_test_scaled = imputer.transform(X_test_scaled)

# Define the classifiers
clf1 = RandomForestClassifier(n_estimators=100, random_state=42)
clf2 = LogisticRegression(random_state=42, solver='liblinear', penalty='l1')
clf3 = KNeighborsClassifier(n_neighbors=5)
clf4 = SVC(probability=True, random_state=42)
```

```python
eclf = VotingClassifier(
    estimators=[('rf', clf1), ('lr', clf2), ('knn', clf3), ('svm', clf4)],
    voting='soft'
)
```

```python
eclf.fit(X_train_val_scaled, y_train_val)
```

```python
# Evaluate the model on the test set
y_test_pred = eclf.predict(X_test_scaled)
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f'Test Accuracy: {test_accuracy:.4f}')
```

```python
def run_for_all(pcap_file):
    packets=rdpcap(pcap_file)
    with open(output_csv_file, 'a', newline='') as csvfile:
        fieldnames = ['Timestamp', 'Start Timestamp', 'TLS Handshake', 'Source IP', 'Destination IP', 'Port', 'Retrain','Label']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        if 'netflix' in pcap_file or 'youtube' in pcap_file:
            label = 'video_stream'
        elif 'skype' in pcap_file:
            label = 'messaging'
        elif 'sftp' in pcap_file:
            label = 'file_transfer'
        elif 'regular_browsing' in pcap_file:
            label = 'regular_browsing'
        else:
            print(pcap_file)
        
        #print(label,pcap_file, len(packets))
        first_packet = True
        count = 0
        # Iterate through packets and write relevant info to the CSV
        for packet in packets[:100]:
            count = count + 1
            #print(count, pcap_file)
            if IP in packet:
                timestamp = packet.time
                if first_packet:
                    start_timestamp = packet.time
                    first_packet = False
                tls_handshake = is_tls_handshake(packet)
                src_ip = packet[IP].src
                dest_ip = packet[IP].dst
                port = packet[IP].sport if TCP in packet or UDP in packet else ''
                payload=payload_to_decimal(packet)
                output_from_last_hidden_layer=retrain(payload)
                
                writer.writerow({
                    'Timestamp': timestamp,
                    'Start Timestamp' : start_timestamp,
                    'TLS Handshake': tls_handshake,
                    'Source IP': src_ip,
                    'Destination IP': dest_ip,
                    'Port': port, 
                    'Retrain':output_from_last_hidden_layer,
                    'Label': label,
                })
```

```python
from tensorflow.keras.models import load_model
model = load_model("neural_net_payload.h5")
```

```python
def is_tls_handshake(packet):
    if packet.haslayer(TCP) and packet[TCP].dport == 443:
        try:
            raw = packet[TCP].load
            if raw and raw.startswith(b'\x16\x03'):
                # TLS Handshake type
                return 'Yes'
        except Exception as e:
            pass
    return 'No'
```

```python
dataset_path = 'input_dataset'

categorized_filenames = {
    'video_stream': [],
    'messaging': [],
    'file_transfer': [],
    'regular_browsing': []
}
```

```python
def payload_to_decimal(packet):
    decimal_payload = []
    if Raw in packet:
        payload = packet[Raw].load
        decimal_payload = [int(b) for b in payload]
    return decimal_payload
```

```python
def retrain(decimal_payload):
    intermediate_layer_model = Model(inputs=model.input, outputs=model.layers[2].output)
    max_length = 20
    input_data = decimal_payload[:20] if len(decimal_payload) > 20 else decimal_payload
    single_input = pad_sequences([decimal_payload], maxlen=max_length, padding='post')
    output_from_last_hidden_layer = intermediate_layer_model.predict(single_input)
    return output_from_last_hidden_layer
```
